```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from skimage.io import imread, imshow
from skimage import filters
```

```python
image1 = imread('UPLOADED_IMAGE.jpg', as_gray=True)
imshow(image1);
```

```python
image2 = imread('Skyfall_1.jpg', as_gray=True)
imshow(image2);
```

```python
image3 = imread('Skyfall_2.jpg', as_gray=True)
imshow(image3);
```

```python
ed_sobel1 = filters.sobel(image1)
imshow(ed_sobel1, cmap='gray');
```

```python
ed_sobel2 = filters.sobel(image2)
imshow(ed_sobel2, cmap='gray');
```

```python
ed_sobel3 = filters.sobel(image3)
imshow(ed_sobel3, cmap='gray');
```

```python
a = np.array(ed_sobel1)
b = np.array(ed_sobel2)
c = np.array(ed_sobel3)


diff1 =  np.mean( a != b )
diff2 =  np.mean( b != c )
diff3 =  np.mean( c != a )

test = np.mean( a != a )
```

```python
print(diff1)
print(diff2)
print(diff3)

print(test)
```
